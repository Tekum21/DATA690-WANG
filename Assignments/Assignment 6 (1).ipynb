### 2).Using Histogram to display the spread of some selected variables.

```python
fig,axis=plt.subplots(2,2,figsize=(12,16))
A=Df["Total Number of School Districts"].values
C=Df["Total Number of Public Schools"].values
E=Df["Total Students  "].values
N=Df["Total Number Operational School Districts "].values

#plot 1
ax=sns.distplot(A,color="blue",ax=axis[0,0],label="Hist.of Total Number of School Districts")
#plot 2
ax=sns.distplot(E,color="#33EEFF",ax=axis[1,0],label="Total Students  ")
#plot 3
ax=sns.distplot(C,color="green",ax=axis[1,1],label="Total Number of Public Schools")
#plot 4
ax=sns.distplot(N,color="red",ax=axis[0,1],label="Total Number Operational School Districts ")
```

###### Using scatter plot to investigate the relationship between:
 * Total Number Operational Schools & Total Students with category = Full-Time Equivalent (FTE) Teachers

```python
#Total Number Operational Schools & Total Students with category = Full-Time Equivalent (FTE) Teachers
fig,ax=plt.subplots(figsize=(8,8))
ax.set_xlabel("Total Number Operational Schools ",color="red")
ax.set_ylabel("Total Students  ",color="red")
ax=sns.scatterplot(x="Total Number Operational Schools ", y="Total Students  ",hue="Full-Time Equivalent (FTE) Teachers  ",data=Df)
ax.grid(True)
```

#### Remarks:
* There are fewer schools with large number of students, with most schools having < 1800000 students.
* The number of full time teachers increases with the size of the student population.

### 3). Using Bar plots to investigate some selected variables.

```python
# Ploting a line plot of x="Total Students  "& y="Total Number Operational Schools.
fig,ax=plt.subplots(figsize=(8,8))
ax=sns.lineplot(x="Total Students  ",y="Total Number Operational Schools ",data=Df)
ax.set_title("Line plt of Total Students VsTotal Number Operational Schools ",color='red',fontsize=14)
plt.xlabel("Total Students  ",color="red",fontsize=14)
plt.ylabel("Total Number Operational Schools ",color="green")
plt.show()
```

```python
ax=sns.lmplot(x="Total Students  ",y="Total Number Operational Schools ",data=Df,height=8,line_kws={'color':"green"},scatter_kws={"color":"#FF33E9"})
plt.xlabel("Total Students  ",color="red",fontsize=14)
plt.ylabel("Total Number Operational Schools ",color="red",fontsize=14)
plt.grid()
# lm plot is a scatter plot with a regression line showing the relationship between Total students and Number of operational schools.
# A linear Regression is a Predictive Statistical Method for modeling a relationship between x(independent varialbe) & y(dependent variable)
```

##### The regression line helps one to predict the No of operational school relative to th student pop.

### 4).Heat Map for some selected variables of interest.

```python
#First we are going to create a dataset that we will display on the heat map-we will use a correlation dataset
#.corr() is used to find all the pairwise correlation of all the columns of the dataframe. Any null values are automatically left out.
#the .corr() compairs the values of the variables between 0-1.
# closser to -1(negative corr), +1(+ corr)
#The closer to 1,or -1 the better.As one variable increse, the other also tends to increase/decrease.
```

```python
pc=Df[["Black Students ","White Students ","Hispanic Students "]].corr()
pc
```

```python
# Displaying Heat Map.
sns.heatmap(pc,annot=True,annot_kws={"size":30})
```

### Remarks:
* We see a stronger corrolation between White & Black students(0.76)
* An average corr between Hispanic & White students(0.67)
