```python
import random

for i in range(120):
    
    print("*", end="")
# to start a new line:
print("\n")
    
# stating our working range  
for i in range(10,30):
# to put the ** at the start    
    print("*", end="")
    
    for i in range(10,30):
        print(" ",random.randrange(10,30)," ",end="")
        
    print("*",end="")
#to put in the ** at the end
    print("\n")    
for i in range(120):
    print("*",end="")
print("\n")
```
