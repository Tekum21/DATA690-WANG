```python
# reimport the data with above 12 columns (ignore all other variables)
df = df[['UNITID','INSTNM','CITY','STABBR','ZIP','ADM_RATE','UGDS','TUITIONFEE_IN','MN_EARN_WNE_P10']]
```

```python
df
```

```python
# Find out the 10 colleges that are hardest to get in (top 10)
hardest_college = df[['INSTNM','ADM_RATE','TUITIONFEE_IN','MN_EARN_WNE_P10']]
hardest_colleges = hardest_college.sort_values('ADM_RATE').head(10)
#OR
hardest_colleges=(10,["ADM_RATE"])
```

```python
# 10. Find out the average in-state tuition of top 10
import statistics
hardest_colleges = hardest_colleges.fillna(0)
#we fill in zeros for those colleges that have value for in-state tuition, without which the calculation willgive us Nan as the result.
statistics.mean(hardest_colleges['TUITIONFEE_IN'])
```

```python
# Find out the average earnings of graduates from the top 10
statistics.mean(hardest_colleges['MN_EARN_WNE_P10'])
```

```python
# Find out the 10 colleges that are easiest to get in (bottom 10)
easiest_college = df[['INSTNM','ADM_RATE','TUITIONFEE_IN','MN_EARN_WNE_P10']]
easiest_colleges = hardest_college.sort_values('ADM_RATE').tail(10)
```

```python
# Find out the average in-state tuition of the bottom 10
easiest_colleges = easiest_colleges.fillna(0)
statistics.mean(easiest_colleges['TUITIONFEE_IN'])
```

```python
# Find out the average earnings of graduates from the bottom 10
statistics.mean(easiest_colleges['MN_EARN_WNE_P10'])
```
