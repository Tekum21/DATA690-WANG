```python
"""Standardized testing has been a cornerstone of the educational system for a long time.
Standardized testing measures a student’s progress throughout their education journey. According to Henry (2007), 
a standardized test is “an examination made up of uniform items which can be replicated, typically by asking
short multiple choice questions which can be easily and cheaply scored by machine”. The use of
standardized testing have expanded in the United States since 2002 after the passage of the No Child Left Behind Act by President George W. Bush,
causing some observers to state that standardized testing is causing a “teaching to the test”
mentality, which undermines America’s ability to produce innovators, imagination, and critical thinkers.
Others state that these standardized exams are a fair and objective way to measure a
student’s abilities.
Besides the debate on whether standardized testing is a good indicator of student progress, 
there is also debate on whether certain factors; teachers role, family support, school environment, peer relationship and the most important factor is the socioeconomic status of parents 
have influnece on a student’s test score and genaral performance. School districts have tried to manipulate
and change certain factors of the educational system to accommodate for these influences,
with the goal of subpressing the negative impacts associated with students test scores. 
The poject below is design to investigate how students' performance in a standardized math, reading and writing test is affected by,their gender, parental level of education, lunch and test preparation course.
It also try to investigate relationships between these variables.  """
#Libraries used here include,pandas,numpy,matplot and seaborn for visualization.
```

```python
#step1: importing the libraries we shall be using.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Step2:Importing our dataset in the form of a csv file.
df=pd.read_csv("Students-Performance.csv")
```

```python
#Step3: We read all the headings in our dataset.
df.columns
```

```python
#Step4: We find out how many rows and columns our dataset contains.
df.shape
```

```python
#Step5: We select our working headings by droping that which is not important to our studies.
df=df.drop(columns=["race/ethnicity"])
df
```

```python
# Adding a column to display each students' total score in the three test.
df["Total Score "]=df.iloc[:,4:].sum(axis=1)
df
```

```python
#We try to find out some statistical values of our dataset.
df.describe()
```

```python
## We use a histogram to display or investigate students' general performance in reading.
bins=[25,50,75,100]
df.hist("reading score",bins=bins,figsize=(8,8))
Av_Reading=69.1
color="red"
plt.axvline(Av_Reading,color=color,label="Av_Reading")
plt.legend()
plt.show()
```

```python
# We use a histogram to display or investigate students' general performance in writing.
bins=[25,50,75,100]
df.hist("writing score",bins=bins,figsize=(8,8))
Av_Writing=68.1
color="red"
plt.axvline(Av_Writing,color=color,label="Av_Writing")
plt.legend()
plt.show()
```

```python
# We use a histogram to display or investigate students' general performance in math.
bins=[25,50,75,100]
df.hist("math score",bins=bins,figsize=(8,8))
Av_Math=66.1
color="red"
plt.axvline(Av_Math,color=color,label="Av_Math")
plt.legend()
plt.show()
```

```python
#From the three displayed  histogram above, we can see that more than half of the students scored above the mean marks in all three test.
```

```python
#Displaying a scatter plot for math and reading scores.
fig, ax = plt.subplots(figsize=(8,8))  
ax.scatter(df['math score'],df['reading score'])
ax.set_title('reading score Vs math score', fontsize=10,color="red")
ax.set_xlabel('reading score', fontsize=10,color='red')
ax.set_ylabel('math score', fontsize=10,color='red')
    
```

```python
#Displaying a scatter plot of reading and writing scores.
fig, ax=plt.subplots(figsize=(8,8))
ax.scatter(df['reading score'],df['writing score'])
ax.set_title("reading score Vs writing score", fontsize=10,color="red")
ax.set_xlabel('reading score',fontsize=10,color='red')
ax.set_ylabel('writing score',fontsize=10,color='red')
```

```python
#Displaying a scatter plot of math and reading scores
fig, ax=plt.subplots(figsize=(8,8))
ax.scatter(df['math score'],df['writing score'])
ax.set_title('math score Vs writing score',fontsize=10, color='red')
ax.set_xlabel("math score",fontsize=10, color='red')
ax.set_ylabel('writing score',fontsize=10,color='red')
```

```python
# From the three scatter plots above showing the relationships between math & reading, reading & writing, and finally writing & math,
#we see taht there is a positive correlation between all three scores,  showing the strogest in correlation between writing & math
#This strong positive correlation indicates that, most students who perform well in any of the test, equally performs well the others.
```

```python
#Investigating the realationship between reading & math score based on students' gender using a bar chart.
fig, ax=plt.subplots(figsize=(16,16))
sns.barplot(x=df["reading score"],y=df["math score"],hue=df['gender'],ci=None)
```

```python
# From the above plot, we can see students who perform well in math also do perform well in reading, with male students performance better in math.
```

```python
# Lets investigate the distribution of students' performance in math using seaborn-boxplot
fig, ax=plt.subplots(figsize=(10,10))
z=df['math score'].values
sns.boxplot(z)
print(df["math score"].median())
#Notes;
#The line in the middle represents the meadian(66.0)
# The box in the middle shows the start of Q1 (25th percentile) and the end is at Q3 (75th percentile)
#The whiskers (left-right) shows the minimum and maximum quartiles.
#The dots to the left show "outliers"
```

```python
#We explore our dataset and rank students' performance in the different test, begining with math.
df1=df.sort_values("math score")
df1
```

```python
#We find out the bottom 250 math scores of our dataset.
a=df1.head(250)
a
```

```python
"""Using scatter plot to investigate if there exist a relationship between students' performance in math and writing based on parents level of eduaction."""
#Using seaborn
fig,ax=plt.subplots(figsize=(8,8))
ax=sns.scatterplot(x='math score',y='writing score',hue='parental level of education',style="parental level of education",data=a)
#Notes: form the plot below, we can the that there is no direct relationship between students' performance and parents' level of education.
```

```python
#Investigating the relationship between writing & reading using their math scores as an indicator.
fig, ax= plt.subplots(figsize=(8,8))
ax=sns.scatterplot(x="reading score",y='writing score',hue='gender',style="gender",size="math score",sizes=(20,200),data=a)
#Notes:From the plot below, we can see that as students' performance generally increases in writing & reading with females performing better, also those who generally perform well in reading & writng also do perform well in math as indicated by the increase in the sizes of the dots.
```

```python
"""Using scatter plot to investigate if there exist a relationship between students' performance in math and writing based on test preparation course."""
#Using seaborn
fig,ax=plt.subplots(figsize=(8,8))
ax=sns.scatterplot(x='math score',y='writing score',hue='test preparation course',style="test preparation course",data=a)
ax.grid(True)
#Notes: form the plot below, we can most students who scored above 60 in both test completed their preparatory course.
```

```python
#Investigating how students' performance in both math & reading is related to their lunch type.
fig,ax=plt.subplots(figsize=(8,8))
ax=sns.scatterplot(x="math score",y="reading score",hue="lunch",style="lunch",data=a)
ax.grid(True)
#The plot below shows that those students who scored above 60 in both math & reading take but standard lunch.
```

```python
#Displaying the relationship between students' performance in math & reading using lmplot.
ax=sns.lmplot(x="math score",y="reading score",height= 8,data=a)
```

```python
#Displaying the relationship between students' performance in math & writing based on test preparation using lmplot
ax=sns.lmplot(x='math score',y='writing score',hue='test preparation course',height=8,data=a, ci=False)
#From the plot, we see that students who completed thier test prep course scored better in both test.
```

```python
#Displaying the relationship students' performance in math & writing based on test preparation using lmplot on two pearate plots.
ax=sns.lmplot(x='math score',y='writing score',col="test preparation course",height=5,data=a, ci=False)
#From the plot, we see that students who completed thier test prep course scored better in both test as indicated by a slightly higher regression line..
```

```python
#Displaying the relationship between students' performance in math & writing based on gender using lmplot
ax=sns.lmplot(x='math score',y='writing score',col='gender',height=5,data=a, ci=False)
#From the plot, we see that female students performed better in writing than their male counterparts.Also more male students instead did well in math than writing. 
```
